```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming your data is in a pandas DataFrame called 'df'
# with columns 'Year' and 'Smoking rate (%)'

# Define the given equation
def custom_equation(x):
    return 1.6449661493638383e+32 * (0.9652 ** x)

# Generate x-values (years) for the equation
x_range = np.linspace(df['Year'].min(), df['Year'].max(), 100)  # Adjust 100 as needed

# Calculate y-values using the equation
y_equation = custom_equation(x_range)

# Plot the smoking rates and the equation
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

plt.plot(df['Year'], df['Smoking rate (%)'], label='Smoking Rates', marker='o', color='blue')  # Smoking rate data
plt.plot(x_range, y_equation, label='y = 1.6449661493638383e+32 * 0.9652^x', color='red', linewidth=2)  # Given equation

plt.xlabel('Year')
plt.ylabel('Smoking Rate (%)')
plt.title('Smoking Rates and Given Equation Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
